## Removing Outlier Values with Modified Z score

```python
sns.boxplot(df.MntRegularProds)
```

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s - median)
    MAD = np.median(diff)
    return MAD

MAD = get_mad(df.MntRegularProds)
median = np.median(df.MntRegularProds)
MAD, median
```

```python
def get_modified_z_score(x, median, MAD):
    return 0.6745 * (x - median) / MAD
```

```python
df["mod_z_score"] = df.MntRegularProds.apply(lambda x: get_modified_z_score(x, median, MAD))
```

```python
df.customer
```

```python
new_df = df[df.mod_z_score < 3.5]
```

```python
sns.boxplot(new_df.MntRegularProds)
```

```python
new_df.describe()
```

```python
sns.boxplot(new_df.Income)
```

```python
new_df["modified_z_score2"] = new_df.Income.apply(lambda x: get_modified_z_score(x, np.median(new_df.Income), get_mad(new_df.Income)))
```

```python
new_df[new_df.modified_z_score2 > 2.5]
```

```python
new_df = new_df[new_df.modified_z_score2 < 2.5]
```

```python
sns.boxplot(new_df.Income)
```
